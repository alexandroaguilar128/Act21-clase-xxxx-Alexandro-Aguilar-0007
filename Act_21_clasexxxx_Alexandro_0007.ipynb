```python
# Definimos la clase Medico
class Medico:
    # El método __init__ es el constructor. Aquí definimos los atributos.
    def __init__(self, bata_blanca, identificacion, instrumento_clinico):
        # Atributos (Datos)
        self.bata_blanca = bata_blanca
        self.identificacion = identificacion
        self.instrumento_clinico = instrumento_clinico

    # Funciones (Acciones)
    def diagnosticar(self):
        print(f"El médico con ID {self.identificacion} está analizando síntomas para dar un diagnóstico.")

    def tratar(self):
        print(f"El médico está usando su {self.instrumento_clinico} para tratar al paciente.")

    def educar(self):
        print("El médico está dando una charla sobre prevención de enfermedades.")

# --- Creación de los Objetos ---

# Objeto 1: Medico1
medico1 = Medico("Larga de seda", "MED-101", "Estetoscopio")

# Objeto 2: Medico2
medico2 = Medico("Corta de algodón", "MED-202", "Termómetro infrarrojo")

# --- Uso de los Objetos ---

print("--- Datos del Médico 1 ---")
print(f"Bata: {medico1.bata_blanca}")
print(f"ID: {medico1.identificacion}")
print(f"Instrumento: {medico1.instrumento_clinico}")
medico1.diagnosticar()
medico1.tratar()

print("\n--- Datos del Médico 2 ---")
print(f"Bata: {medico2.bata_blanca}")
print(f"ID: {medico2.identificacion}")
print(f"Instrumento: {medico2.instrumento_clinico}")
medico2.educar()
print("Programa terminado")
print("Alexandro Aguilar NL = 1")
```
